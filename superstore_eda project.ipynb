# Literature Review

The retail industry generates vast amounts of transactional data through customer purchases, product sales, inventory movements,
and promotional activities. Analyzing this data is crucial for understanding business performance, customer preferences, and
profitability patterns. Exploratory Data Analysis (EDA) serves as a foundational step in extracting meaningful insights from 
retail datasets by identifying trends, anomalies, relationships, and patterns within the data.

Several studies have highlighted the importance of data analytics in improving retail decision-making. According to Tang (2025),
retail organizations can leverage sales data to identify customer purchasing behavior, forecast demand, and optimize inventory
management. The study emphasized that sales analytics enables businesses to improve operational efficiency and profitability
through data-driven strategies.

Customer segmentation has emerged as another significant area of retail analytics. Research conducted by Doğan et al. 
(2018) demonstrated that analyzing customer transaction data using RFM (Recency, Frequency, Monetary) models and clustering 
techniques helps businesses identify high-value customer groups and design targeted marketing campaigns. Such analytical 
approaches contribute to improved customer retention and revenue generation.

Furthermore, Kaewpradit (2025) highlighted the role of retail data analysis in uncovering purchasing patterns and customer behavior. 
The study found that customer-centric analytics assists organizations in developing personalized business strategies and enhancing 
overall customer satisfaction. These findings reinforce the importance of understanding sales trends and customer segments before
implementing strategic business decisions.

Product profitability analysis has also received significant attention in retail research. Various studies suggest that high sales 
volumes do not always translate into high profits. Factors such as discounts, operational costs, and product-specific margins can
substantially affect profitability. Consequently, category-level and sub-category-level analysis is widely used to identify 
profit-driving and loss-making products, enabling organizations to optimize pricing and promotional strategies.

Regional performance analysis represents another critical aspect of retail business intelligence. Previous research indicates that
geographical variations influence customer demand, product preferences, and sales performance. Evaluating sales and profit across
different regions helps organizations allocate resources effectively and identify growth opportunities in underperforming markets.

The existing literature consistently demonstrates the value of Exploratory Data Analysis as a preliminary step in retail analytics
projects. EDA facilitates data understanding, quality assessment, trend identification, and hypothesis generation before advanced
statistical or machine learning techniques are applied. Visualization techniques such as histograms, boxplots, correlation heatmaps,
scatter plots, and pair plots are frequently employed to reveal hidden patterns and support decision-making.

In the present study, Exploratory Data Analysis is performed on the Superstore retail dataset to investigate sales performance,
profitability, customer segments, product categories, sub-categories, discounts, and regional variations. The analysis aims
to identify key business drivers, uncover profitability challenges, and generate actionable insights that can support strategic 
decision-making. By applying descriptive statistics and visualization techniques, this study contributes to a comprehensive
understanding of retail business performance and aligns with established practices in retail data analytics research.

```python
# A. Yearly Sales Trend

df['Order Date'] = pd.to_datetime(df['Order Date'])

yearly_sales = df.groupby(df['Order Date'].dt.year)['Sales'].sum()

plt.figure(figsize=(8,5))
yearly_sales.plot(marker='o')

plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

 Insights  
  
Sales performance changes over time, indicating business growth or decline across years.  
The yearly trend helps identify long-term business performance patterns.  
Consistent growth in sales suggests increasing market demand and customer acquisition.  
Fluctuations in annual sales may indicate external market influences or operational changes.  
Year-over-year analysis provides valuable insights into overall business expansion.

```python
#  Yearly Profit Trend

yearly_profit = df.groupby(
    df['Order Date'].dt.year
)['Profit'].sum()

yearly_profit.plot(marker='o', figsize=(8,4))
plt.title('Yearly Profit Trend')
plt.ylabel('Profit')
plt.show()
```

Insights  
  
2017 generated the highest profit, reflecting strong business growth and improved financial performance over time.  
The overall upward trend in yearly profit indicates that the company successfully expanded its operations and customer base.  
Consistent profit growth across years demonstrates the sustainability of the business model.  
The increasing profitability suggests that revenue growth was accompanied by effective cost and margin management.  
The positive long-term trend reflects the company's ability to create increasing value over time.

Understanding monthly demand patterns can support inventory and marketing planning.

```python
#  Monthly Profit Trend

monthly_profit = df.groupby(
    df['Order Date'].dt.month_name()
)['Profit'].sum().reindex([
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
])

monthly_profit.plot(marker='o', figsize=(10,5))
plt.title('Monthly Profit Trend')
plt.ylabel('Profit')
plt.show()
```

```python
# C. Seasonal Sales Analysis 

season_map = {
    12:'Winter', 1:'Winter', 2:'Winter',
    3:'Spring', 4:'Spring', 5:'Spring',
    6:'Summer', 7:'Summer', 8:'Summer',
    9:'Fall', 10:'Fall', 11:'Fall'
}

season_sales = df.groupby(
    df['Order Date'].dt.month.map(season_map)
)['Sales'].sum()

season_sales = season_sales.reindex(
    ['Winter','Spring','Summer','Fall']
)

plt.figure(figsize=(8,5))
sns.barplot(
    x=season_sales.index,
    y=season_sales.values
)

plt.title('Seasonal Sales Analysis')
plt.xlabel('Season')
plt.ylabel('Total Sales')
plt.show()
```

 Insights  
  
Fall recorded the highest sales, indicating strong customer demand during the latter part of the year and contributing significantly
to overall revenue generation.    
Spring generated comparatively lower sales, suggesting reduced purchasing activity during this period compared to other seasons.    
Sales increased steadily from the early seasons toward Fall, highlighting the presence of a clear seasonal sales pattern.    
The noticeable variation in sales across seasons demonstrates that customer purchasing behavior is influenced by seasonal factors.    
The strong performance in Fall suggests that inventory planning, marketing campaigns, and resource allocation should be prioritized 
during high-demand seasons to maximize revenue opportunities.
